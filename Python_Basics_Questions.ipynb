# Python Basics: Beginner Questions
This notebook covers basic Python concepts with code examples.

### 1. Integer vs Float
**Integer** is a whole number (no decimal), while **Float** is a number with a decimal point.
```python
x = 10       # Integer
y = 3.14     # Float
```

### 2. Declaring a String Variable
```python
name = "Alice"
greeting = "Hello, " + name
print(greeting)
```

### 3. What is a List?
A **list** is a collection of items which is mutable (can be changed).
```python
fruits = ["apple", "banana", "cherry"]
print(fruits[1])  # Access element
fruits[1] = "orange"  # Modify element
print(fruits)
```

### 4. Tuple vs List
A **tuple** is immutable (cannot be changed), while a **list** is mutable.
```python
colors = ("red", "green", "blue")  # Tuple
animals = ["cat", "dog"]           # List
# Use tuple for fixed data, list for dynamic data
```

### 5. Dictionary Creation and Access
```python
person = {"name": "John", "age": 25}
print(person["name"])
```

### 6. Type Casting
```python
# Implicit
a = 5
b = 2.0
c = a + b  # c is float

# Explicit
x = "10"
y = int(x)
print(y + 5)
```

### 7. Convert String to Integer and Perform Arithmetic
```python
num_str = "20"
num = int(num_str)
print(num + 10)
```

### 8. String Concatenation
```python
first = "Good"
second = "Morning"
combined = first + " " + second
print(combined)
```

### 9. Indexing in Lists and Strings
```python
text = "Python"
print(text[0])  # 'P'

colors = ["red", "green", "blue"]
print(colors[-1])  # 'blue'
```

### 10. Student Age Dictionary
```python
students = {
    "Alice": 20,
    "Bob": 22,
    "Charlie": 19
}
print("Bob's age:", students["Bob"])
```
